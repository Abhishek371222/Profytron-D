## 2. Run ESLint and Capture Report
Run ESLint from the terminal and collect the output summary of errors and warnings.

import subprocess, pathlib
root = pathlib.Path(r'c:\Users\abhis\Downloads\ye final karo')
report = root / 'eslint-report.json'
cmd = ['pnpm', 'exec', 'eslint', '.', '--ext', '.ts,.tsx', '--format', 'json']
print('Running', cmd)
result = subprocess.run(cmd, cwd=root, capture_output=True, text=True)
print('returncode=', result.returncode)
print('stderr sample:', result.stderr[:1000])
report.write_text(result.stdout, encoding='utf-16')
print('Report written to', report)

## 3. Apply Auto-fix for Fixable Issues
